```python
import json
import requests
from time import sleep
from tqdm import tqdm
import os
import datetime

def dumpFile(new_dir_path, new_filename, new_file_content, mode='w'):
    os.makedirs(new_dir_path, exist_ok=True)
    with open(os.path.join(new_dir_path, new_filename), mode) as f:
        json.dump(new_file_content, f, ensure_ascii=False, sort_keys=True, separators=(',', ': '))
        f.close()

def getData():
    res=[]
    #Êñ∞„Åó„Åè„Éá„Ç£„É¨„ÇØ„Éà„É™„Çí‰Ωú„Å£„Å¶„Åì„Åì„Å´„Éá„Éº„Çø„Çí‰øùÂ≠ò„Åô„Çã
    new_dir='./'+'problem_data'
    
    filename='problem_list.json'
    url="https://codeforces.com/api/problemset.problems"
    response=requests.get(url)
    json_data=response.json()

    dumpFile(new_dir,filename,json_data)
    return json_data

getData()
```
